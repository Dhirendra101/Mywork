```python
from sentence_transformers import SentenceTransformer
from elasticsearch import Elasticsearch
```

```python
sentence_transformer = SentenceTransformer("bert-base-nli-mean-tokens")
```

```python
questions = [

    "How do I improve my English speaking? ",

    "How does the ban on 500 and 1000 rupee notes helps to identify black money? ",

    "What should I do to earn money online? ",

    "How can changing 500 and 1000 rupee notes end the black money in India? ",

    "How do I improve my English language? "

]

question_embeddings = sentence_transformer.encode(questions)
```

```python
es_client = Elasticsearch("localhost:9200",timeout=1000)
print(es_client)
INDEX_NAME = "faq_bot_index"

EMBEDDING_DIMS = 768
```

```python
mappingRequest = """{
          "settings" : {
            "index" : {
              "number_of_shards" : 3, 
                "number_of_replicas" : 1
            }
          },
          "mappings" : {
            "properties" : {
                "Id" : { "type": "integer" },
                "embedding" : { "type": "dense_vector","dims": EMBEDDING_DIMS },
                "question" : { "type": "text" },
                "answer" : { "type": "text" }
            }
        }
        }"""
```

```python
def create_index() -> None:

    es_client.indices.delete(index=INDEX_NAME, ignore=404)

    es_client.indices.create(

        index=INDEX_NAME,

        ignore=400,

        body={
          "settings" : {
            "index" : {
              "number_of_shards" : 3, 
                "number_of_replicas" : 1
            }
          },
          "mappings" : {
            "properties" : {
                "Id" : { "type": "integer" },
                "embedding" : { "type": "dense_vector","dims": EMBEDDING_DIMS },
                "question" : { "type": "text" },
                "answer" : { "type": "text" }
            }
        }
        })

create_index()
```

```python
def index_qa_pairs(qa_pairs):
    for qa_pair in qa_pairs:
        question = qa_pair["question"]
        
        answer = qa_pair["answer"]

        embedding = sentence_transformer.encode(question).tolist()
        print(embedding)
        print("===============================================================================")
        
        Id=qa_pair["id"]

        data = {
            "Id":Id,

            "question": question,

            "embedding": embedding,

            "answer": answer,

        }

        es_client.index(

            index=INDEX_NAME,

            body=data

        )

    

```

```python
qa_pairs = [{
    "id":1,
    "question": "How do I improve my English speaking? ",

    "answer": "Speak more",

},{
    "id":2,
    "question": "What should I do to earn money online? ",

    "answer": "Learn machine learning",

},{
    "id":3,
    "question": "How can I improve my skills? ",

    "answer": "More practice",

}]

```

```python
index_qa_pairs(qa_pairs)
```

```python
ENCODER_BOOST = 10
def query_question(question: str, top_n: int=10):
    embedding = sentence_transformer.encode(question).tolist()
    es_result = es_client.search(
        index=INDEX_NAME,
        body={
            "from": 0,
            "size": top_n,
            "_source": ["question", "answer","embedding"],
            "query": {
                "script_score": {
                    "query": {"match_all": {}},
                      "script": {
                      "source": "(cosineSimilarity(params.query_vector, 'embedding') + 1.0)* params.encoder_boost + _score",
                      "params": {"query_vector": embedding,
                                 "encoder_boost": ENCODER_BOOST,}
    }
  }
            }
        }
    )
    hits = es_result["hits"]["hits"]
#     clean_result = []
#     for hit in hits:
#         clean_result.append({
#             "question": item["_source"]["question"],
#             "answer": item["_source"]["answer"],
#             "score": item["_score"],
#         })
        
    return hits


hits=query_question("")
```
